```python
import numpy as np
from spektral.datasets import qm9
```

```python
A, X, E, y = qm9.load_data(return_type='numpy',
                           nf_keys='atomic_num',
                           ef_keys='type',
                           self_loops=True,
                           amount=None)  # Set to None to train on whole dataset
```

```python
y
```

```python
tasks = list(y.columns)[1:]

corr_dict = dict()
for i in range(len(tasks)):
    for j in range(i+1):
        corrcoef_ij = np.corrcoef(y[[tasks[i], tasks[j]]], rowvar=False)[0, 1]
        print(tasks[i], tasks[j], corrcoef_ij)
        if tasks[i] not in corr_dict:
            corr_dict[tasks[i]] = dict()
        if tasks[j] not in corr_dict:
            corr_dict[tasks[j]] = dict()
        corr_dict[tasks[i]][tasks[j]] = corrcoef_ij
        corr_dict[tasks[j]][tasks[i]] = corrcoef_ij
```

```python
with open('corrcoef.csv', 'w') as f:
    print(end=',', file=f)
    for task_j in tasks:
        print(task_j, end=',', file=f)
    print(file=f)
    for task_i in tasks:
        print(task_i, end=',', file=f)
        for task_j in tasks:
            print(corr_dict[task_i][task_j], end=',', file=f)
        print(file=f)
```

```python
def cluster_by_correlations(corr_dict, tasks, threshold):
    adj_tasks = dict()
    threshold = 0.5
    for task_i in tasks:
        adj_tasks[task_i] = set()
        for task_j in tasks:
            if abs(corr_dict[task_i][task_j]) > threshold:
                adj_tasks[task_i].add(task_j)
        for task_j in tasks:
            not_done = True
            while not_done:
                not_done = False
                tasks_to_add = set()
                for adj_task in adj_tasks[task_i]:
                    if abs(corr_dict[task_j][adj_task] > threshold) and task_j not in adj_tasks[task_i]:
                        tasks_to_add.add(task_j)
                adj_tasks[task_i] |= tasks_to_add
                if len(tasks_to_add) > 0:
                    not_done = True

    for task in tasks:
        sorted_list = tuple(sorted(list(adj_tasks[task])))
        adj_tasks[task] = sorted_list

    task_collections = set(adj_tasks.values())
    return task_collections
```
